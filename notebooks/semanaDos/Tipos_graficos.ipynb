# Tipos de gráficas

En esta sección del curso, exploraremos las técnicas visuales disponibles en Matplotlib para representar datos por pares, o Pairwise data. 

Desde el  gráfico de líneas hasta herramientas como fill_between y stackplot, descubriremos cómo cada tipo de gráfico nos brinda una perspectiva única sobre la relación entre dos variables cuantitativas. 


```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
f = '../data/Cuernavaca_1dia_comas.csv'
cuerna = pd.read_csv(f,index_col=0,parse_dates=True)
cuerna.head()
```

```python
fig, ax = plt.subplots()

ax.plot(cuerna.To)
ax.set_title('plot')
```

```python
fig, ax =  plt.subplots()

ax.scatter(cuerna.index,cuerna.To)
ax.set_title('Scatter')
```

```python
fig, ax = plt.subplots()
# f1 = parse("2012-01-01")
# f2 = f1 + pd.Timedelta("10H")

ax.bar(cuerna.index,cuerna.To)
# ax.bar(cuerna.index,cuerna.To,width=1/25)

ax.set_title('Bars')


# ax.set_xlim(f1,f2)
```

```python
fig, ax = plt.subplots()

ax.stem(cuerna.To)
```

```python
fig, ax = plt.subplots()

ax.fill_between(cuerna.index,cuerna.To.mean(),cuerna.To)
ax.plot(cuerna.To,'r-')
```

```python
fig, ax = plt.subplots()

ax.stairs(cuerna.To)
```
